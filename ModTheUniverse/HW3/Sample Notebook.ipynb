# Problem Set 3 Starter Code


This is a quick example of how to read Kepler data using kplr (as described in the problem, there are other ways to read the data).

```python
import kplr
import numpy as np

# Find the target KOI.
client = kplr.API()
koi = client.koi(97.01)

# Get a list of light curve datasets.
lcs = koi.get_light_curves(short_cadence=False)

# Open the first dataset and read it
f = lcs[0].open()
hdu_data = f[1].data
time = hdu_data["time"]  # get the time of each observation
flux = hdu_data["sap_flux"] # get the flux
flux_err = hdu_data["sap_flux_err"] # get the error in the flux
f.close()
```

```python
print(time, flux, flux_err)
```
